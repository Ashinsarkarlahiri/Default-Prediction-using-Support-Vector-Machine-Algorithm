```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.svm import SVC
from sklearn.metrics import accuracy_score, classification_report
from sklearn.preprocessing import StandardScaler
```

```python
default = pd.read_csv('default_data.csv')
```

```python
default=default.dropna()
```

```python
X = default [["balance_time","mat_time", "gdp_time","uer_time","interest_rate_time","hpi_time","FICO_orig_time"]]
Y= default ["default_time"]
```

```python
X
```

```python
Y
```

```python
x_train,x_test,y_train,y_test= train_test_split(X,Y,train_size=0.8,random_state=6) 
```

```python
# standardize the features (imp for svm):
scaler = StandardScaler()
x_train = scaler.fit_transform(x_train)
x_test = scaler.fit_transform(x_test)

# make a support vector machine classifier model:
svm_classifier = SVC(kernel = 'linear', C =1.0 , random_state = 45)

# train the classifier on the training set:
svm_classifier.fit(x_train,y_train)

# make predictions on the testing set:
y_pred = svm_classifier.predict(x_test)

# Evaluate model performance:
accuracy = accuracy_score(y_test,y_pred)
print(f'Accuracy: {accuracy}')

# You can also print other metrics through classification report:
print(classification_report(y_test,y_pred))
```
